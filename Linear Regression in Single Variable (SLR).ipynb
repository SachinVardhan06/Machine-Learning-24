```python
import pandas as pd
import seaborn as sns
from sklearn import linear_model
```

```python
df = pd.read_csv("age_premium.csv")
df
```

```python
sns.lmplot(x='age',y='premium',data=df)
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['age']],df[['premium']])
```

```python
reg.predict([[21]])
```

```python
reg.predict([[85]])
```

```python
# Slope of Line
reg.coef_
```

```python
# Intercept of Line
reg.intercept_
```
